## ADD a column base_salary. 
####  => if career_Level = L1, then base_salary= 9999
####  => if career_Level = L2, then base_salary= 8888
####  => if career_Level = L3, then base_salary= 7777
####  => if career_Level = L4, then base_salary= 6666



```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\HP\Documents\subho\EmployeeDatabase.csv')
```

```python
df.head()
```

```python
def assignSal(lvl):
    if lvl == 'L1':
        return 9999
    elif lvl == 'L2':
        return 8888
    elif lvl == 'L3':
        return 7777
    elif lvl == 'L4':
        return 6666
```

```python
df['base_salary'] = df.Careel_Level.map(assignSal)
```

```python
df.head()
```

## if skill is any one of  'pyspark', 'aws', 'scala', then they are eligible for next gen

```python
def checkNextGen(skl):
    if skl in ['pyspark', 'aws', 'scala']:
        return 'yes'
    else:
        return 'no'
```

```python
df['Next_Gen_eligibility'] = df['Skills'].map(checkNextGen)
```

```python
df.head()
```

## calculate number of working days i.e, today's date - joining date

```python
import datetime
def calExpDays(dt):
    dtFmt = datetime.datetime.strptime(dt, '%d%b%Y')
    date_diff = datetime.datetime.now() - dtFmt
    if date_diff.days > 0:
        return date_diff.days
    return 0
```

```python
df['WorkingDays'] = df['Joining_Date'].map(calExpDays)
```

```python
df.head()
```

## calculate actual salary = base_salary * 3.14 + WorkingDays * Num_of_project / 365

```python
df['actual_salary'] = df['base_salary'] * 3.14 + df['WorkingDays'] * df['Num_of_project']/365
```

```python
df.head()
```

```python
df['exp'] = df['actual_salary'] * 0.4
df['worth'] = df['actual_salary']-df['exp']
```

```python
df.head()
```

```python
def checkPromo(wdays, Next_Gen_eligibility):
    if wdays > 1000 and Next_Gen_eligibility=='yes':
        return 'yes'
    elif wdays > 5000:
        return 'yes'
    else:
        return 'no'
```

```python
df.apply(lambda x: checkPromo(x.WorkingDays, x.Next_Gen_eligibility), axis =1)
```
